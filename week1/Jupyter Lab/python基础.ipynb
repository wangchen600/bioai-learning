6. 函数定义

```python
def calculate_gc_percent(sequence):
    """计算DNA序列的GC百分比"""
    sequence = sequence.upper()
    gc_percent = ((sequence.count('G') + sequence.count('C')) / len(sequence)) * 100
    return gc_percent

test_seq = "ATGCGCGCAT"
print(f"'{test_seq}'的GC含量: {calculate_gc_percent(test_seq):.1f}%")
```

练习题

```python
"""
分析DNA序列的基本信息
参数:
    sequence: DNA序列字符串（只包含ATCG）
返回:
    字典: 包含长度、A/T/C/G数量、GC含量
"""
def analyze_dna(sequence):
    dna_data_dict = {}
    sequence = sequence.upper()
    dna_len = len(sequence)
    a = sequence.count('A')
    t = sequence.count('T')
    c = sequence.count('C')
    g = sequence.count('G')
    gc_percent = (g + c) / dna_len * 100
    dna_data_dict['lenth'] = dna_len
    dna_data_dict['A'] = a
    dna_data_dict['T'] = t
    dna_data_dict['C'] = c
    dna_data_dict['G'] = g
    dna_data_dict['GC含量'] = gc_percent
    return dna_data_dict

test_seq = "ATGCGTACGTAGCTAGCTAGCATGCTAGCTAGC"
result = analyze_dna(test_seq)
print(f"'{test_seq}'的GC含量: {result['GC含量']:.1f}%")
```
