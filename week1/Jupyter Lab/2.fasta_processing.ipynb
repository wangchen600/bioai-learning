1.FASTA文件处理

```python
from Bio import SeqIO
import pandas as pd
```

1.1 读取FASTA文件中的所有序列

```python
sequences = []
file_path = "/week1/output/lactobacillus_16s.fasta"

for record in SeqIO.parse(file_path,'fasta'):
    sequences.append({
        "id": record.id,
        "description": record.description,
        "sequence": str(record.seq),
        "length": len(record.seq)})

print(f"共读取 {len(sequences)} 条序列")
df = pd.DataFrame(sequences)
print(df)
```

2. 批量序列分析

```python
from Bio.SeqUtils import gc_fraction

# 为每条序列计算GC含量
for record in SeqIO.parse(file_path, "fasta"):
    seq_str = str(record.seq)
    gc = gc_fraction(seq_str) * 100
    print(f"{record.id}: 长度={len(seq_str)}, GC含量={gc:.1f}%")
```

3. 结果保存为CSV

```python
results = []
for record in SeqIO.parse(file_path, "fasta"):
    seq_str = str(record.seq)
    results.append({
        "id": record.id,
        "length": len(seq_str),
        "GC_percent": gc_fraction(seq_str) * 100,
        "A_count": seq_str.count('A'),
        "T_count": seq_str.count('T'),
        "C_count": seq_str.count('C'),
        "G_count": seq_str.count('G')
    })

rna_df = pd.DataFrame(results)
rna_df.to_csv("D:\\python\\学习\\week1\\output\\sequence_analysis_results.csv", index=False)
print("结果已保存到 sequence_analysis_results.csv")
print(rna_df)
```
