1. Seq对象操作

1.1.导入Biopython的Seq模块

```python
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
```

1.2.创建一个DNA序列对象

```python
dna_seq = Seq("ATGCGTACGTAGCTAGCTAGC")
print(f"DNA序列: {dna_seq}")
print(f"序列类型: {type(dna_seq)}")
```

1.3.基本序列操作

```python
dna_seq = Seq("ATGCGTACGTAGCTAGCTAGC")
print(f"长度: {len(dna_seq)}")
print(f"反向序列: {dna_seq[::-1]}")
print(f"互补序列: {dna_seq.complement()}")
print(f"反向互补序列: {dna_seq.reverse_complement()}")
```

1.4.转录（DNA → RNA）

```python
dna_seq = Seq("ATGCGTACGTAGCTAGCTAGC")
rna_seq = dna_seq.complement().transcribe()
print(f"RNA序列: {rna_seq}")
```

1.5.翻译时遇到终止密码子停止

```python
dna_seq = Seq("ATGCGTACGTAGCTAGCTAGC")
protein_seq_stop = dna_seq.translate(to_stop=True)
print(f"翻译到终止密码子: {protein_seq_stop}")
```

1.6.使用不同的遗传密码表

```python
# table=2 表示脊椎动物线粒体密码子
dna_seq = Seq("ATGCGTACGTAGCTAGCTAGC")
protein_mito = dna_seq.translate(table=2)
print(f"线粒体密码子翻译: {protein_mito}")
```

1.7.GC含量计算

```python
dna_seq = Seq("ATGCGTACGTAGCTAGCTAGC")
gc_content = gc_fraction(dna_seq) *100
print(f"GC含量: {gc_content}%")
```

2.序列文件格式转换

```python
#2.1 手动创建一个SeqRecord对象
seq = Seq("ATGCGTACGTAGCTAGCTAGC")
record = SeqRecord(seq,
                   id="test_gene_001",
                   name="test_gene",
                   description="A test gene sequence for learning Biopython")
#2.2 写入FASTA文件
with open("output/test.fasta", "w") as output_handle:
    SeqIO.write( record,output_handle, "fasta")

print("FASTA文件已生成")
#2.3 读取FASTA文件
with open("output/test.fasta", "r") as input_handle:
    for record in SeqIO.parse(input_handle, "fasta"):
        print(f"ID: {record.id}")
        print(f"描述: {record.description}")
        if len(record) > 50:
            print(f"序列: {record.seq[:50]}...")
        else:
            print(f"序列: {record.seq[:50]}")
        print(f"长度: {len(record.seq)}")
```
